```python
#Plot with seaborn main command sns.countplot for sales['Branch'] 
sns.countplot(sales['City']) 
plt.xlabel('Branch Name')
plt.ylabel('Count')
plt.title('Most busy branch')
A,B,C = sales.Branch.value_counts()
```

```python
#Question 1.	Which Supermarket branch is the most profitable?
# We use the gross income to determine the profit of branches. 

profit_branch = sales.groupby('City').agg([np.sum, np.median, np.mean]) 
profit_branch[['gross income']].T
```

```python
# Naypyitaw has gross income of 5265.17 higher than Mandalay gross income of 5057.03 and Yangon gross income of 5057.16
```

```python
#Plot the supermarket branch profit 
profit_branch_plot = sns.boxplot(x="City", y="gross income", data=sales)
```

```python
#GENDER ANALYSIS

men = sales['Gender'].value_counts(ascending = True) # Count each gender values 
men
```

```python
sns.countplot(x="Gender", data =sales).set_title("Gender count")
```

```python
# Product sales by gender
plt.figure(dpi=100)
sns.countplot(y ='Product line', hue = "Gender", data = sales)
```

```python
#2. Does gender affect income?

#Gross income by gender
profit_branch = sales.groupby('Gender').agg([np.sum, np.median, np.mean]) 
profit_branch[['gross income']].T
```

```python
# Total gross income made by males = 7384.94 and female = 7994.43. And gender determine income. females generate more income than males, hence more female sales means more income 
```

```python
#plot of gross income by gender.
sns.catplot(x='Gender',y='gross income',data=sales, kind = 'point', aspect = 4 )
```

```python
# Gross income by gender at different branches
sns.catplot(x='Gender',y='gross income', hue ='Branch', data=sales, kind = 'violin', aspect = 3, palette= "Set2" )
```

```python
#PRODUCT ANALYSIS

# Quatity of product line items sold. Electronic accessories sold highest followd by Food and beverages. Health and beauty sold least
productline_qty = sales.groupby('Product line').agg([np.sum, np.median, np.mean]) 
productline_qty[['Quantity']].T
```

```python
#Question 3.	Which of the product line performs best in the Supermarket?

# Product vs incomes. Product making bringing more income are home & lifstyle, food & beverages and sports and travel
plt.figure(dpi=150)
sns.barplot(x ='Product line', y = 'gross income', data = sales)
plt.xticks(rotation=75)
plt.show()

productline_income = sales.groupby('Product line').agg([np.sum, np.median, np.mean]) 
productline_income[['gross income']].T
```

```python
# In terms of gross income, Home and lifestyle product line averagely brings the highest income followed by Sports and travel
```

```python
#4.	What attribute had significant importance in sales profit?
# First, we find the correlation among the attributes and income gross income (=sales profit)
```

```python
#use heatmap to show for correlation 
plt.figure(dpi=120)
sns.heatmap(np.round(sales.corr(), 2), annot=True)
plt.show(), 
```

```python
# From the correlation, 5% tax fee for customer buying, Total price including tax, Cost of goods sold attributes correlates highly with gross income. 
```
